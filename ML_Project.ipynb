# Split the data

Here I made both random split and stratified split to compare the perfomance between them but in my running code I implement the stratified split because it's performs better .

# Random Split

```python
# Random Split
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(forset_cover,test_size=0.2,random_state=21)
```

```python
train_set.shape
```

```python
test_set.shape
```

# Stratified Split

```python

import numpy as np 
forset_cover["Elevation cat"]=pd.cut(forset_cover['Elevation'],
                                     bins=[1800,2050,2300,2550,2800,3050,3300,3550,3800,np.inf],
                                     labels=[1,2,3,4,5,6,7,8,9])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=23)
```

```python
for train_index, test_index in split.split(forset_cover, forset_cover["Elevation cat"]):
    train_set = forset_cover.loc[train_index]
    test_set = forset_cover.loc[test_index]
```

# Prepare the Data

```python
# prepare train set
forset_train_prepare=train_set.drop(columns=['Cover_Type'])
target_train=train_set['Cover_Type']

# prepare the test set
forset_test_prepare=test_set.drop(columns=["Cover_Type"])
target_test=test_set["Cover_Type"]

```

```python
target_train
```

```python
target_test
```

```python
# make standardization for train set and test set
from sklearn.preprocessing import StandardScaler
standerd=StandardScaler()
forset_train_prepare_std=forset_train_prepare
forset_train_prepare_std=standerd.fit_transform(forset_train_prepare_std.values)

forset_test_prepare_std=forset_test_prepare
forset_test_prepare_std=standerd.fit_transform(forset_test_prepare_std.values)
```

```python
forset_train_prepare_std.shape
```

# Build The Models

# Logistic Regression

```python
# Training the model
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(forset_train_prepare_std,target_train)
```

```python
# extract the accuracy
accuracy_lr = log.score(forset_test_prepare_std, target_test)
print("Logistic Regression Accuracy:", accuracy_lr*100)

# Here we can see that the accuracy is too bad 
```

```python
# Calculate the Accuracy,Precision,Recall,f1_score,Confusion Matrix

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

predict=log.predict(forset_test_prepare_std)

accuracy =accuracy_score(target_test,predict)

precision=precision_score(target_test,predict,average='weighted')

recall=recall_score(target_test,predict,average='weighted')

f1=f1_score(target_test,predict,average='weighted')

confusion=confusion_matrix(target_test,predict)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(confusion)
```

```python
# Implement hyperparameter tuning
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],    
    'penalty': ['l1', 'l2'],                       
    'solver': ['liblinear', 'saga'],                
    'class_weight': [None, 'balanced'],                          
    'tol': [1e-4, 1e-3],                          
}

grid_search = GridSearchCV(log, param_grid, cv=2, n_jobs=-1, verbose=1)

grid_search.fit(forset_train_prepare_std, target_train)

print("Best parameters:", grid_search.best_params_)

# Can't see the result because it takes much time to run 
```

# Decision Tree Model

```python
# Here I will use the  data without scaling because it's decision tree model

# Train the model
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(forset_train_prepare,target_train)

# Test the model using test set and display the accuracy
accuracy_dtc=dtc.score(forset_test_prepare,target_test)
print("Dections tree Accuracy =",accuracy_dtc*100)
```

```python
from sklearn.model_selection import cross_val_score

scores =cross_val_score(dtc,forset_train_prepare,target_train,scoring='accuracy',cv=10 ) 
print("Scores:", scores*100)
# Here I make cross validation Just to showing the accuracy in defferent division of trainingset
```

```python
# Calculate the Accuracy,Precision,Recall,f1_score,Confusion Matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

predict=dtc.predict(forset_test_prepare)

accuracy =accuracy_score(target_test,predict)

precision=precision_score(target_test,predict,average='weighted')

recall=recall_score(target_test,predict,average='weighted')

f1=f1_score(target_test,predict,average='weighted')

confusion=confusion_matrix(target_test,predict)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(confusion)
```

```python
# Let's make hyperparameter tuning for decision tree classifier 

from sklearn.model_selection import  GridSearchCV

param_grid = {
    'criterion': ['gini', 'entropy'],           
    'splitter': ['best', 'random'],                         
    'max_depth': [3, 5, 10, None],                          
    'min_samples_split': [2, 5, 10],                         
    'min_samples_leaf': [1, 2, 4],                          
    'min_weight_fraction_leaf': [0.0, 0.01, 0.1],          
    'max_features': [None, 'sqrt', 'log2'],                
    
}

grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=2,scoring='accuracy' ,n_jobs=-1, verbose=1)
grid_search.fit(forset_train_prepare, target_train)
best_params = grid_search.best_params_

print("Best Hyperparameters:", best_params)


```

```python
best_params
```

```python
# After hyperparameter tuning train the model again using the best parmeters and look at the new accuracy


from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(**best_params)
dtc.fit(forset_train_prepare,target_train)


accuracy_dtc=dtc.score(forset_test_prepare,target_test)
print("Dections tree Accuracy =",accuracy_dtc*100)

# Here you can see the accuracy increased a little bit
```

```python
# Again after hyperparameter tuning 
# Calculate the Accuracy,Precision,Recall,f1_score,Confusion Matrix

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

predict=dtc.predict(forset_test_prepare)

accuracy =accuracy_score(target_test,predict)

precision=precision_score(target_test,predict,average='weighted')

recall=recall_score(target_test,predict,average='weighted')

f1=f1_score(target_test,predict,average='weighted')

confusion=confusion_matrix(target_test,predict)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:")
print(confusion)
```
